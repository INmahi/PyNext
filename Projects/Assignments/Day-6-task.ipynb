```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "C:/Users/DELL/PyNext/Datasets/titanic - titanic6.csv"
titanic = pd.read_csv(file_path)
```

```python
titanic = titanic.drop("Unnamed: 0",axis=1,errors='ignore') #this dataset is the cleaned and missing value handled version, exported from day5 task. exporting caused an extra index column to appear. so i dropped it first. used gpt for it
titanic.head()
titanic.shape
```

```python
titanic['date_of_birth'] = pd.date_range(start='1910-01-01', periods=len(titanic), freq='D')

titanic.head()
```

```python
filtered_titanic = titanic[titanic['date_of_birth'] < '1910-06-01']
filtered_titanic
```

```python
youngest = titanic.sort_values(by='date_of_birth', ascending=False)
youngest.head()
```

```python
oldest = titanic.sort_values(by='date_of_birth', ascending=True)
oldest.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic['Survived_label'] = titanic['Survived'].map({0: 'No', 1: 'Yes'})

sns.countplot(data=titanic, x='Survived_label')
plt.title('Count of Survivors vs Non-Survivors')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
titanic['Sex'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightpink'])
plt.title('Percentage of Males vs Females')
plt.ylabel('')
plt.show()
```

```python
plt.hist(titanic['Fare'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=titanic, y='Fare', color='lightcoral')
plt.title('Box Plot of Fare Distribution')
plt.ylabel('Fare')
plt.show()
```
