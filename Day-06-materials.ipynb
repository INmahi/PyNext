# Matplotlib

```python
import matplotlib.pyplot as plt
month = ['Jan', 'Feb', 'Mar', 'Apr']
income = [14, 34, 34, 23]
expense = [12,32, 23, 45]


plt.plot(month, income, color = 'green', linewidth = 5,
        linestyle = '-.')
plt.xlabel('Month')
plt.ylabel('Income')
plt.title('Line plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
month = ['Jan', 'Feb', 'Mar', 'Apr']
income = [14, 34, 34, 23]
expense = [12,32, 23, 45]

fig,ax = plt.subplots(figsize = (5,2))
ax.plot(month, income, color = 'green', linewidth = 5,
        linestyle = '-.')
ax.set_xlabel('Month')
ax.set_ylabel('Income')
ax.set_title('Line plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
month = ['Jan', 'Feb', 'Mar', 'Apr']
income = [14, 34, 34, 23]
expense = [12,32, 23, 45]

fig,(ax1, ax2) = plt.subplots(2,1,figsize = (8,6))
ax1.plot(month, income, color = 'green', linewidth = 5,
        linestyle = '-.')
ax1.set_xlabel('Month')
ax1.set_ylabel('Income')
ax1.set_title('Line plot')

ax2.plot(month, expense, color = 'red', linewidth = 5,
        linestyle = '--')
ax2.set_xlabel('Month')
ax2.set_ylabel('Expense')
ax2.set_title('Line plot')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
month = ['Jan', 'Feb', 'Mar', 'Apr']
income = [14, 34, 34, 23]
expense = [12,32, 23, 45]
df = {
    'month':month,
    'income':income,
    'expense':expense
}
df = pd.DataFrame(df)
print(df)
```

```python
sns.set_style('whitegrid')
sns.lineplot(data = df, x = 'month', y = 'income')
```

```python
sns.barplot(data = df, x = 'month', y = 'income')
```

```python
sns.histplot(data = df, x = 'income')
plt.xlabel('income')
plt.show()
```

```python
sns.boxplot(data = df, y = 'income')
plt.xlabel('income')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,(ax1, ax2) = plt.subplots(2,1,figsize = (8,6))
sns.lineplot(data = df, x = 'month', y = 'income', 
         ax = ax1, color = 'green')
ax1.set_xlabel('Month')
ax1.set_ylabel('Income')
ax1.set_title('Line plot')

sns.lineplot(data = df, x = 'month', y = 'expense', 
         ax = ax2, color = 'red')
ax2.set_xlabel('Month')
ax2.set_ylabel('Expense')
ax2.set_title('Line plot')
plt.tight_layout()

plt.savefig('plt.pdf', format = 'pdf')
plt.show()
```
