```python
# clean categories
def normalize_dept(s):
    s=s.astype(str).str.strip().str.lower()
    mapping = {
        "cse":"CSE","c.s.e":'CSE',
        "eee":"EEE","E.E.E":'EEE',"eleng":"EEE",'e.e.e':"EEE",
        "mat":"MAT","Mmathematics":"MAT",
        "sta":"STA","statistics":'STA',
        'phy':"PHY","physics":"PHY",
        'me':"MEE",'mee':"MEE",
        'ipe':'IPE','industrial production':"IPE"
        
    }

    return s.replace(mapping).str.upper()
```

```python
df["dept_clean"] = normalize_dept(df["dept"])
df
```

```python
df.groupby("dept_clean")["student_id"].count()
```

```python
df.groupby("gender")["student_id"].count()
```

```python
def normalize_gender(s):
    s=s.astype(str).str.strip().str.lower()
    mapping = {
        "f":"Female","female":'Female',
        "m":"Male","male":'Male',
        "other":"Other"          
    }

    return s.replace(mapping).str.capitalize()
```

```python
df["gender_clean"] = normalize_gender(df["gender"])
```

```python
df.groupby("gender_clean")["student_id"].count()
```

```python
df["height_cm_clean"] = df["height_cm"].where((df["height_cm"]>=130)&(df["height_cm"]<=210))
```

```python
df["height_cm_clean"].max()
```

```python
df["weight_kg_clean"] = df["weight_kg"].where((df["weight_kg"]>=30)&(df["weight_kg"]<=120))
```

```python
df["weight_kg_clean"].max()
```

```python
df["BMI"] = df["weight_kg_clean"]/(df["height_cm_clean"]/100)**2
df
```

```python
att_cols = [c for c in df.columns if c.startswith("att_T")]
```

```python
att_cols
```

```python
df["Attendance_rate"] = df[att_cols].mean(axis=1)*100
df["Attendance_rate"] 
```

```python
ontime_cols = [c for c in df.columns if c.startswith("ontime_A")]
df["ontime_rates"] = df[ontime_cols].mean(axis=1)*100
df["ontime_rates"]
```

```python
marks_cols = [c for c in df.columns if c.startswith("marks_A")]
df["total_marks"] = df[marks_cols].sum(axis=1)
df["mark_rate"] = (df["total_marks"]/700)*100
```

```python
df["mark_rate"]
```

```python
df["Avg_marks"] = df[marks_cols].mean(axis=1,skipna=True)
df["Avg_marks"]
```

```python
df['final_score'] = 0.7*df['Attendance_rate'] + 0.3*df["ontime_rates"] + df["Avg_marks"]
```
