```python
import pandas as pd

data = {
    'timestamp': ['01-01-2023 10:00 AM',
                  '01-01-2023 11:00 AM',
                  '02-01-2023 08:30 PM'],
    'value': [10, 15, 20]
}
df = pd.DataFrame(data)

print(type(df["timestamp"]))

time = pd.to_datetime(df['timestamp'], format="%d-%m-%Y %I:%M %p")

print(type(time))

```

```python
print(pd.date_range(start='2025-08-01',end='2025-08-29',freq = '3D'))
print(pd.date_range(start='2025-08-01',end='2025-09-29',freq = 'W')) #starts from sunday
print(pd.date_range(start='2025-08-01',end='2025-09-29',freq = pd.DateOffset(weeks = 1)))  #starts from start
print(pd.date_range(start='2025-08-01',end='2025-09-29',freq = "ME"))  #starts from 31
print(pd.date_range(start='2025-08-01',end='2025-09-29',freq = pd.DateOffset(months =1)))  #starts from 1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "month": ["Jan", "Feb", "Mar", "Apr"],
    "Edu_ex": [150, 300, 100, 400],
    "Food_ex": [500, 200, 600, 350],
    "Entertainment_ex": [100, 450, 200, 500]
}

df = pd.DataFrame(data)

df[["Edu_ex", "Food_ex", "Entertainment_ex"]].sum().plot(kind="pie",autopct="%.2f%%",colors=["#ff9999", "#66b3ff", "#99ff99"],ylabel="")

plt.title("Expense Distribution", fontsize=14, fontweight="bold")
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
income = [45, 78, 32, 67, 54, 89, 21, 63, 48, 75, 56, 92]
expense = [45, 78, 32, 67, 54, 89, 21, 63, 48, 75, 56, 92]
expense = [30, 50, 25, 40, 35, 60, 15, 45, 28, 55, 40, 70]

fig,ax = plt.subplots(figsize = (15,3))
ax.plot(months,income,color="red",linewidth="4",linestyle="--")
ax.set_xlabel('Month')
ax.set_xlabel('Month')
ax.set_title('Line Plot')
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
income = [45, 78, 32, 67, 54, 89, 21, 63, 48, 75, 56, 92]
expense = [30, 50, 25, 40, 35, 60, 15, 45, 28, 55, 40, 70]

fig,(ax1,ax2) = plt.subplots(1,2,figsize = (15,5))

ax1.plot(months,income,color="green",linewidth="4",linestyle="-")
ax1.set_xlabel('Month')
ax1.set_xlabel('Income')
ax1.set_title('Line Plot')

ax2.plot(months,expense,color="red",linewidth="4",linestyle="-")
ax2.set_xlabel('Month')
ax2.set_xlabel('Expense')
ax2.set_title('Line Plot')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
income = [45, 78, 32, 67, 54, 89, 21, 63, 48, 75, 56, 92]
expense = [30, 50, 25, 40, 35, 60, 15, 45, 28, 55, 40, 70]


dict = {
    'month': months,
    'income': income,
    'expense': expense
}
df = pd.DataFrame(dict)
sns.set_style('whitegrid')
sns.lineplot(data=df,x='month',y='income')
```

```python
sns.barplot(data=df,x='month',y='income')
```

```python
sns.histplot(data=df,x='income')
plt.xlabel('income')
```

```python
import matplotlib.pyplot as pit

import seaborn as sns

fig,(ax1, ax2)= plt.subplots (2,1,figsize = (8,6))

sns.lineplot(data = df, x= 'month', y= 'income', 
             ax = ax1, color = 'green')

ax1.set_xlabel('Month')

ax1.set_ylabel('Income')

ax1.set_title('Line plot')

sns.lineplot(data= df, x = 'month', y = "expense", ax = ax2, color = 'red')

ax2.set_xlabel('Month')

ax2.set_ylabel('Expense')

ax2.set_title('Line plot')

plt.tight_layout()

plt.show()
```
